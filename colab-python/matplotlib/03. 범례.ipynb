```python
# import 구문
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
# matplotlib 에서 해당 폰트를 사용할 수 있도록 등록

for f_path in fm.findSystemFonts(fontpaths=None, fontext='ttf'):
  fm.fontManager.addfont(f_path)

 # matplotlib 기본 글꼴 설정
plt.rc("font", family="NanumMyeongjo")
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y, label="데이터")
plt.legend();
```

- 두 개의 선을 하나의 차트에 그릴 때

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 4, 7]
```

```python
plt.plot(x1,y1, label="data1", color="red")
plt.plot(x2,y2, label="data2", color="blue")
plt.legend();
```

```python
plt.plot(x1,y1, color="red")
plt.plot(x2,y2, color="blue")
plt.legend(["data1","data2"]);
```

><b>2) 범례 폰트 사이즈 지정</b>

```python
plt.plot(x1,y1, color="red")
plt.plot(x2,y2, color="blue")
plt.legend(["data1","data2"], fontsize=20);
```

><b>3) 범례 행 수 지정</b>

```python
plt.plot(x1,y1, color="red")
plt.plot(x2,y2, color="blue")
plt.legend(["data1","data2"], ncol=2);
```

- [학생문제] : data1(x : 1, 2, 3; y : 1, 2, 3) 과 data2(x : 1, 2, 3; y : 1, 100, 200) 을 그래프로 출력하시오.

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 100, 200]
```

```python
plt.plot(x1,y1, color="blue")
plt.plot(x2,y2, color="red");
```
