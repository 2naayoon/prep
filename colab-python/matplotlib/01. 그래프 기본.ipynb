### 기본 차트 생성

><b>리스트를 그래프로 나타내기</b>

```python
#

x = [10, 20, 30, 40]

# ; : 로그 출력 중지
plt.plot(x);

g = plt.title("그래프")
```

```python
#

x = [1,2,3,4]
y = [12,43,25,15]

g = plt.plot(x, y)
```

><b>파이썬 딕셔너리를 그래프로 나타내기</b>

```python
data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}

plt.plot('data_x','data_y', data = data_dict);
```
