### subplots()

```python
fig,ax  = plt.subplots(2,2, figsize=(15,10))
fig.suptitle("여러 그래프 작성")

# 첫번째 그래프 작성(0,0)
ax[0,0].bar(df['이름'],df['국어'],label='국어점수')
ax[0,0].legend()
ax[0,0].set_title('학생별 국어점수')
ax[0,0].grid(alpha=0.2)

# 두번째 그래프 작성(0,1)
ax[0,1].plot(df['이름'],df['영어'],label='영어점수')
ax[0,1].plot(df['이름'],df['수학'],label='수학점수')
ax[0,1].legend()
ax[0,1].set_title('학생별 국어/수학 점수')

# 세번째 그래프 작성(1,0)
ax[1,0].barh(df['이름'],df['키'],label='키')
ax[1,0].legend()
ax[1,0].set_title('학생별 키')

# 네번째 그래프 작성(1,1)
ax[1,1].plot(df['이름'],df['사회'],color="g",alpha=0.5);
```
