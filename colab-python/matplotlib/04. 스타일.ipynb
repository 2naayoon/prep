```python
# import 구문
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
# matplotlib 에서 해당 폰트를 사용할 수 있도록 등록

for f_path in fm.findSystemFonts(fontpaths=None, fontext='ttf'):
  fm.fontManager.addfont(f_path)

 # matplotlib 기본 글꼴 설정
plt.rc("font", family="NanumGothic")
```

```python
x = [1,2,3]
y = [2,4,8]
```

#### 1. 선 굵기 : linewidth

```python
plt.plot(x,y,linewidth=10)
```

#### 2. 마커
- https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html


```python
plt.plot(x,y,marker="o");
```

```python
# 마커 사이즈 변경

plt.plot(x,y,marker="o", markersize=10);
```

```python
plt.plot([4,5,6], marker="H")
plt.plot([3,4,5], marker="d")
plt.plot([2,3,4], marker="x")
plt.plot([1,2,3], marker=11)
plt.plot([0,1,2], marker="$Z$");
```

```python
# markeredgecolor : 마커 테두리 색상
# markerfacecolor : 마커 색상

plt.plot(x,y, marker='o',markersize=20,markeredgecolor="red",markerfacecolor="yellow");
```

#### 3. 선 모양 : linestyle(https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)

```python
plt.plot(x,y, linewidth=5, linestyle=":");
```

```python
plt.plot(x,y, linewidth=5, linestyle="--");
```

```python
plt.plot(x,y, linewidth=5, linestyle="-.");
```

#### 4. 선 색상 : color

- https://matplotlib.org/stable/api/colors_api.html#module-matplotlib.colors

```python
plt.plot(x,y, linewidth=5, color="r");
# r = red / b = blue / g = green
```

```python
plt.plot(x,y, linewidth=5, color="b");
```

```python
plt.plot(x,y, linewidth=5, color="g");
```

#### 5. 포맷 - fmt = '[marker][line][color]'

```python
plt.plot(x,y, 'o');
```

```python
plt.plot(x,y, 'ro');
```

```python
plt.plot(x,y, 'ro--');
```

```python
plt.plot(x,y, 'bv:');
```

```python
plt.plot(x,y, 'go');
```

#### 6. 축약(https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
# mfc : markerfacecolor
# mec : markeredgecolor
# ls : linestyle
# ms : markersize

plt.plot(x,y,marker='o', mfc="r",mec="b",ls=":", ms=10);
```

#### 7. 투명도

```python
plt.plot(x,y,marker='o', mfc="r",mec="b", ms=10, alpha=0.3);
```

#### 8. 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y);
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y);
```

```python
plt.figure(figsize=(5,10), dpi=200)
plt.plot(x,y);
```

#### 9. 배경색

```python
plt.figure(facecolor="yellow")
plt.plot(x,y);
```
